```python
def load_words(src_path='corpus.txt'):
    """Load words from corpus and return the valid word set with its size.
    Return: size, words
    """
    words = []
    with open(src_path, 'r') as f:
        while True:
            w = f.readline()
            if not w: break
            w = w.strip()
            if w.isalpha() and len(w) == 5:
                words.append(w)
    return len(words), words
```

```python
def extend_corpus(src_path, dst_path='corpus.txt'):
    """Append new words to the destination file."""
    src_f = open(src_path, 'r')
    dst_f = open(dst_path, 'a')
    while True:
        w = src_f.readline()
        if not w: break
        dst_f.write(w)
    src_f.close()
    dst_f.close()
```

```python
extend_corpus('test_src.txt', 'test_dst.txt')
```
